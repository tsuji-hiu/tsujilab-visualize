```python
# グラフを書いてみよう

import numpy as np
import matplotlib.pyplot as plt
```

```python
yoko = [1, 2, 3, 4, 5]
tate = [2, 5, 3, 6, 4]

plt.plot(yoko, tate)
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 5, 3, 6, 4]

plt.plot(x, y)
```

```python
# マーカーをつける

x = [1, 2, 3, 4, 5]
y = [2, 5, 3, 6, 4]

plt.plot(x, y, marker='o')
```

```python
# 練習問題1: "M" のグラフ

x = [1, 2, 3, 4, 5]
y = [3, 6, 3, 6, 3]

plt.plot(x, y, marker='o')
```
