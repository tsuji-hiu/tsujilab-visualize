```python
# グラフを書いてみよう

import numpy as np
import matplotlib.pyplot as plt
```

```python
# 練習問題2: 直線のグラフ

x = [0, 1, 2, 3, 4]
y = [0, 3, 6, 9, 12]

plt.plot(x, y, marker='o')
```

```python
# 練習問題2: 直線のグラフ（ndarray）

x = np.array([0, 1, 2, 3, 4])
y = np.array([0, 3, 6, 9, 12])

plt.plot(x, y, marker='o')
```

```python
# 練習問題2: 直線のグラフ（ndarray）

x = np.array([0, 1, 2, 3, 4])
y = 3 * x

plt.plot(x, y, marker='o')
```

```python
# 練習問題2: 直線のグラフ（ndarray）

x = np.linspace(0, 4, 5)
y = 3 * x

plt.plot(x, y, marker='o')
```
